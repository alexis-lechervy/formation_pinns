# Projet de cours : Prédire l'absorption des médicaments à l'aide d'un réseau de neurones informé par la physique

*D'apèrs un sujet de He Li (School of Chemical, Materials, and Biomedical Engineering, University of Georgia)*

## Énoncé du projet

La modélisation mathématique de la pharmacocinétique traite de la vitesse d'assimilation, de transport et d'élimination des médicaments par le système humain au cours du temps et aide à comprendre l'efficacité et la réponse thérapeutique des médicaments dans le corps humain. En particulier, le modèle à compartiments a été utilisé pour étudier le processus d'assimilation des médicaments, dans lequel le corps humain est représenté comme une séquence de compartiments. Le nombre de compartiments dans ce modèle dépend de la nature et du comportement du médicament considéré ainsi que de la complexité du modèle lui-même.

Dans ce projet, nous considérerons un modèle à trois compartiments, ce qui signifie que le corps humain est représenté par trois compartiments, à savoir le tractus gastro-intestinal (tractus GI), le flux sanguin et le tractus urinaire. Nous supposons en outre que le tractus urinaire est un compartiment absorbant uniquement, c'est-à-dire que le médicament entre mais n'est pas éliminé du tractus urinaire. Le médicament est d'abord absorbé dans le tractus gastro-intestinal (tractus GI), puis diffusé dans le flux sanguin, et finalement éliminé du flux sanguin par les reins et le foie vers le tractus urinaire. Ces trois processus sont caractérisés par différentes vitesses en fonction des différents constituants du même médicament.

## Modèles à compartiments

L'un des premiers modèles à compartiments développés par Widmark et al. [4] était un modèle à compartiment ouvert, visant à étudier la concentration d'alcool dans le sang. Edward et al. [7] ont présenté un modèle pharmacocinétique à deux compartiments plus sophistiqué. Plus tard, Koch et al. [6] ont utilisé des modèles à un et deux compartiments pour étudier les modèles d'injection unique et multiple. Khanday et al. [5] ont établi un modèle mathématique pour estimer la distribution des médicaments dans le sang et les tissus en utilisant des méthodes orales et intraveineuses. Récemment, Egbelowo et al. [2] ont tenté de résoudre un modèle pharmacocinétique à un compartiment pour différentes voies d'administration des médicaments.

Pour les besoins du projet, nous travaillerons avec le cas d'une dose unique de médicament. Dans ce modèle, une prise unique de médicament est considérée comme entrant dans le tractus GI, qui est le premier compartiment d'où le médicament se dissout et diffuse dans le flux sanguin, qui est le deuxième compartiment. À partir du flux sanguin, le médicament est éliminé par les reins et le foie vers le tractus urinaire, qui est le compartiment absorbant. L'ensemble du système est décrit par un système de trois équation différentielle ordinaire.

$$
\begin{aligned}
& \frac{d B}{d t}=k_{g} G-k_{b} B, \quad B(0)=0 \\
& \frac{d G}{d t}=-k_{g} G, \quad G(0)=G_{0} \\
& \frac{d U}{d t}=k_{b} B, \quad U(0)=0
\end{aligned}
$$

où $G(0)=G_{0}$ est la quantité initiale de médicament prise, $k_{g}$ est la constante de vitesse ou la constante de proportionnalité. $B(0)=0$ est la quantité initiale de médicament dans le flux sanguin, $k_{b}$ est une autre constante de vitesse ou constante de proportionnalité et $U(0)=0$ est la quantité initiale de médicament dans le tractus urinaire. Ainsi, le modèle à trois compartiments, qui décrit la vitesse de changement du médicament lorsqu'il diffuse à travers le système humain, est donné par les 3 équations ci-dessus. Les valeurs des vitesses auxquelles le médicament diffuse du tractus GI dans le flux sanguin puis est éliminé sont respectivement de $0,72 \, \text{heure}^{-1}$ et $0,15 \, \text{heure}^{-1}$. La quantité initiale de médicament dans le tractus GI est de 0,1 nanogramme. Les valeurs des coefficients ont été prises de Belinda et al [1], où le médicament considéré est l'antibiotique tétracycline. Pour plus de détails, veuillez vous référer à [3].
